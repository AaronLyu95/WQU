## Introduction

In the stock market, investors have access to three main types of stocks: growth stocks, value stocks, and income stocks. Income stocks, also known as dividend stocks, distinguish themselves by their commitment to continuously paying dividends to shareholders. These income stocks can provide investors with a steady income, which seem to be more attractive to those investors who seeking a reliable source of returns. Regular income from dividends provides a buffer against market fluctuations and provides some predictability and financial security. While income stocks are an attractive option for investors seeking stability, it is important to be aware of potential challenges. The value of an investment can be sensitive to changes in interest rates.

## Extract Data

The data was extracted from Yahoo Finance, and time period is from 2020-01-01 to 2024-01-01

```python
start_date = '2020-01-01'
end_date = '2024-01-01'
```

```python
def extract_security(symbol, start_date='2020-01-01', end_date='2024-01-01'):
    ticker = yf.Ticker(symbol)
    df = ticker.history(start = start_date, end = end_date, period='1d')
    df.reset_index(inplace=True)
    df['Date'] = df['Date'].dt.date
    df['symbol'] = symbol
    return df
```

### Verizon Communications Inc. (VZ)  

```python
df_vz = extract_security('VZ', start_date, end_date)
```

## EDA

```python
def compute_daily_return(df, price_name='Close', period=1):
    df['price_with_dividend'] = df[price_name] + df['Dividends']
    df['daily_return'] = df[price_name]/df[price_name].shift(period)-1
    df['daily_return_with_dividends'] = df['price_with_dividend']/df['price_with_dividend'].shift(period)-1
    df.drop(columns=['price_with_dividend'], inplace=True)
    return df
```

### Verizon Communications Inc. (VZ)  

```python
daily_return_vz = compute_daily_return(df_vz, 'Close', 1)
```

### CVR Energy, Inc. (CVI) 

```python
daily_return_cvi = compute_daily_return(df_cvi, 'Close', 1)
```

### The Toronto-Dominion Bank (TD)  

```python
daily_return_td = compute_daily_return(df_td, 'Close', 1)
```

### Statistics about returns

```python
def custom_kurtosis(series):
    return kurtosis(series, fisher=False, nan_policy='omit')
```

```python
df = pd.concat([daily_return_vz, daily_return_cvi, daily_return_td], axis=0)
```

```python
df.head()
```

```python
grouped_data = df.groupby('symbol').agg({
    'daily_return': ['mean', 'std', 'skew', custom_kurtosis],
    'daily_return_with_dividends': ['mean', 'std', 'skew', custom_kurtosis]
})
```
